# Koşullu Seçim (Conditional Selection)

```python
# Veri setinde yaşı 50'den büyük olan kişileri seçmek istediğimde kullandığımız sorgu aşağıdaki gibidir;

df[df["age"] > 50].head()
```

```python
# Yaşı 50'den büyük olan kişileri sayrdırmak istediğimizde;
# Aşağıya dikkat edildiğinde en son oluşturulan df üzerinde bir sütun seçimi yapmamız gerekir.


df[df["age"] > 50]["age"].count()
```

```python
# Yaşı 50'den büyük olan kişilerin class'larını da merak ettiğimizde;

df.loc[df["age"] > 50, ["age", "class"]].head()
```

```python
# Yaşı 50'den büyük olan erkekleri seçmek istediğimizde;
# DİKKAT!!! Birden fazla koşul giriliyorsa koşulların parantez içerisine alınması gerekir.

df.loc[(df["age"] > 50) & (df["sex"] == "male"), ["age", "class"]].head()
```

```python
# Yaşı 50'den büyük olup, embark_town'u "Cherbourg" olan erkekleri seçmek istediğimizde;

df.loc[(df["age"] > 50) 
        & (df["sex"] == "male") 
        & (df["embark_town"] == "Cherbourg"), 
        ["age", "class", "embark_town"]].head()

# Kodun Okunulabilirliğinin de arttırılmasın özen gösterilmelidir.
```

```python
# Yaşı 50'den büyük olup, embark_town'u "Cherbourg" ya da "Southampton" olan erkekleri seçmek istediğimizde;

df.loc[(df["age"] > 50) 
        & (df["sex"] == "male") 
        & ((df["embark_town"] == "Cherbourg") | (df["embark_town"] == "Southampton")), 
        ["age", "class", "embark_town"]].head()
```

# Toplulaştırma ve Gruplama (Aggregation & Grouping)

```python
# - count()
# - first()
# - last()
# - mean()
# - median()
# - min()
# - max()
# - std()
# - var()
# - sum()
# - pivot table

# pivot table haricinde diğer fonksiyonaları groupby ile kullanabiliriz.
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
# insanların yaş ortalaması

df["age"].mean()
```

```python
# Kadın ve erkeklerin yaş ortalamasını almak istersem;

df.groupby("sex")["age"].mean()
```

```python
# Bir diğer metod ise;
# Aşağıda bakıldığı zaman bir sözlük kullanımı gibi bir ifade vardır.

df.groupby("sex").agg({"age": "mean"})
```

```python
# Bu metodun kullanılması tavsiye edilebilir, çünkü birden fazla özet istatistik istenebilir;

df.groupby("sex").agg({"age": ["mean", "sum"]})
```

```python
# Bir başka örnek;

df.groupby("sex").agg({"age": ["mean", "sum"],
                       "survived": "mean"})
```

```python
# Önce cinsiyete göre daha sonra bindikleri limanlara göre yaş ve yaşam kırılımı yapmak istersek;


df.groupby(["sex", "embark_town"]).agg({"age": ["mean"],
                                        "survived": "mean"})
```

```python
# Önce cinsiyete göre sonra bindikleri limanlara daha sonra class'larına göre yaş ve yaşam kırılımı yapmak istersek;


df.groupby(["sex", "embark_town", "class"]).agg({"age": ["mean"],
                                                 "survived": "mean"})
```

```python
# Önce cinsiyete göre sonra bindikleri limanlara daha sonra class'larına göre yaş, yaşam kırılımı ve cinsiyet kırılımı yapmak istersek;


df.groupby(["sex", "embark_town", "class"]).agg({
                                                 "age": ["mean"],
                                                 "survived": "mean",
                                                 "sex": "count"})
```

# Pivot table

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.pivot_table("survived", "sex", "embarked")

# Yukarıda birinci argümana kesişimde neyi görmek istiyorsak onu giriyoruz,
# İkinci argüman olarak satırda/indekste görmek istediğimi argümanı giriyoruz,
# Üçüncü argümanda ise sütunda görmek istediğimiz argümanı giriyoruz.

# Değer olarak ortalama değer gelir, çünkü pivot table'ın ön tanımlı değeri ortalamadır.
```

```python
df.pivot_table("survived", "sex", "embarked", aggfunc="std")

# yukarıdaki pivot table da standart sapma hesaplanmış oldu.
```

```python
df.pivot_table("survived", "sex", ["embarked", "class"])

# Daha sonrasında embarked'i de class kırılımında görmek istersek
```

```python
# Yaşlara göre de kırılım görmek istersek;
# Yaşlar float olduğu için gruplama yapmam gerekecek.
# Bunu yapmanın yolu sayısal değişkeni kategorik değişkene çevirmektir.
# Sayısal bir değişkeni kategorik bir değişkene çevirirken pd.cut() veya pd.qcut() işlemi uygulanır.
# Eğer ki böleceğimiz değişkeni tanıyorsak pd.cut() kullanarak nerelerden bölmesi gerektiğini söyleyebiliriz.
# Ama böleceğimiz değişkeni tanımıyorsak pd.qcut kullanarak direk çeyrekliklerine bölebiliriz.

df["new_age"]=pd.cut(df["age"], [0, 10, 18, 25, 40, 90])
df.head()
```

```python
# Şimdi yaşa göre kırılım işlemini uygularsak;

df.pivot_table("survived", "sex", "new_age")
```

```python
# Yukarıdaki yaş kırılımına bir de class kırılımı eklersek;

df.pivot_table("survived", "sex", ["new_age", "class"])
```

# Apply ve Lambda

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age2"] = df["age"]*2
df["age3"] = df["age"]*5

df.head()
```

```python
# Veri setimizdeki yaş değişkenlerimizi 10'a böleceğimizi düşünelim,

(df["age"]/10).head()
(df["age2"]/10).head()
(df["age3"]/10).head()

# ilk kısmı parantez içerisine almazsak .head() metodunu kullanamayız hata verir!!!
```

```python
# Yukarıdaki işlemi döngü ile yapmak istersek;

for col in df.columns:
    if "age" in col:
        print(col)
```

```python
# Devamında ise;

for col in df.columns:
    if "age" in col:
        print((df[col]/10).head())
```

```python
df

# Yukarıda bakıldığı df'e bakıldığı zaman bir değişiklik olmadığını görüyoruz.
```

```python
# Değişikliğin olabilmesi için;

for col in df.columns:
    if "age" in col:
        df[col]=df[col]/10

df
```

```python
# Bu işlemi apply ve lambda ile aşağıdaki gibi kısa yoldan da yapabiliriz;

df[["age", "age2", "age3"]].apply(lambda x:x/10).head()
```

```python
# Yukarıda age içerenleri elimizle yazdık.
# Eğer ki bularak yapmak isteseydik;

df.loc[:, df.columns.str.contains("age")].apply(lambda x:x/10).head()
```

```python
# Biraz daha zorlaştırmak istersek :)
# Yaş üzerinde bir standartlaştırma işlemi yapalım;  ((x-x.mean())/x.std())

df.loc[:, df.columns.str.contains("age")].apply(lambda a: (a-a.mean())/a.std())
```

```python
# Yukarıda apply'ın içerisinde dışardan bir fonksiyon gönderilebilir mi?

def standart_scaler(col_name):
    return (col_name-col_name.mean())/col_name.std()


df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)
```

```python
# Yukarıdaki işlemleri kaydetmek için ise;

# df.loc[:, ["age","age2","age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)
```

# Birleştirme (join) işlemleri

```python
m=np.random.randint(1,30, size=(5,3))

# Yukarıda m adında 1 ile 30 arasında rasgele sayılarla 5'e 3'lük bir matris oluşturuluyor.

m
```

```python
df1=pd.DataFrame(m, columns=["var1", "var2", "var3"])

# yukarıda m ile sütun isimleri "var1", "var2", "var3" olan bir df oluşturuluyor.

df2=df1+99

print(df1)
print(df2)
```

```python
# Eğer ki yukarıdaki iki df'yi birleştirmek istersem .concat() metodunu kullanırız.

pd.concat([df1, df2])
```

```python
# Yukarıdaki tabloda indeksi aynı olan değerler bulunmaktadır. Bu problemi aşmak için ise;

pd.concat([df1, df2], ignore_index=True)

# Yanyana birleştirme gerçekleştirmek istersem .concat() metodunun içerisine "axis=1" argümanını gönderirim.
```

```python
# Merge ile Birleştirme İşlemleri

df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})

print(df1)
print(df2)
```

```python
pd.merge(df1, df2)

# hangi değişkene göre birleştirmesi gerektiğini vermememize rağmen çalışanlara göre birleştireceğini anladı.
```

```python
# Lakin biz yine de birleştireceği değişkeni vermek istersek;

pd.merge(df1, df2, on="employees")
```

```python
# Amaç: Her çalışanın müdürünün bilgisine erişmek istiyoruz.
df3 = pd.merge(df1, df2)

df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})

pd.merge(df3, df4)
```

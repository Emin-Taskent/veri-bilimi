# MATPLOTLIB

```python
# Kategorik değişken: sütun grafik. countplot bar
# Sayısal değişken: hist, boxplot
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
df.head()
```

```python
# Kategorik Değişken Görselleştirme

df['sex'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Sayısal Değişken Görselleştirme

plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

# Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

```python
plt.plot(x, y)

# Başlık
plt.title("Bu ana başlık")

# X eksenini isimlendirme
plt.xlabel("X ekseni isimlendirmesi")

plt.ylabel("Y ekseni isimlendirmesi")

plt.grid()
plt.show()
```

# Subplots

```python
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 2, 1)        # 1 satırlık 2 sütunluk bir grafik oluşturacağızımı söylüyor ve 1. kısma ilk grafik çıktısını vereceğimi söylüyorum
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 2, 2)        # 1 satırlık 2 sütunluk bir grafik oluşturacağızımı söylüyor ve 2. kısma ilk grafik çıktısını vereceğimi söylüyorum
plt.title("2")
plt.plot(x, y)
plt.show()
```

```python
# 3 grafiği bir satır 3 sütun olarak konumlamak.
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)
plt.title("2")
plt.plot(x, y)

# plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 3)
plt.title("3")
plt.plot(x, y)

plt.show()
```

# SEABORN

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt


df = sns.load_dataset("tips")
df.head()
```

```python
# Kategorik veri görselleştirme

df["sex"].value_counts()
sns.countplot(x=df["sex"], data=df)

# x eksenini ve verinin nereden alındığını .countplot() metodu içerisinde belirtmemiz gerekir.

plt.show()
```

```python
# Sayısal Değişken Görselleştirme

sns.boxplot(x=df["total_bill"])
plt.show()


# Aşaığda pandas kütüphanesi kullanılmak istenirse;


df["total_bill"].hist()
plt.show()


```

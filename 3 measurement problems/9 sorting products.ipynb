### Sorting by Rating, Comment and Purchase

```python
df.head()
```

```python
# Ürünlerimizi; purchase_count, rating ve comment count'a göre sıralama işlemi yapacağız.
# Bunu yaparken de purchase_count, rating ve comment count değişkenlerini standardize hale getirmemiz bizim için daha iyi olacaklatır.
# Yoksa değişkenler arasında uygun bir ağırlıklandırma yapılamaz.
# Bu case için purchase_count değişkeni diğer değişkenlerden daha etkili hale gelmiş olur.

# Bu sebeple standardizasyon işlemi yapmamız gerekir.
# Bu standardizasyon işlemi ise minmaxscaler ile yapılacaktır.

df["purchase_count_scaled"] = MinMaxScaler(feature_range=(1, 5)). \
    fit(df[["purchase_count"]]). \
    transform(df[["purchase_count"]])

# Yukarıdaki ko ile yeni bir sütun oluşturuyoruz, daha sonrasında rating'e göre (en düşük puan 1 en yüksek 5 olduğu için),
# ölçeklendireceğimizi söylüyoruz. Sonrasında "purchase_count" değişkenini fit etmesi gerektiğini ifade ediyoruz.
# En son olarak da değiştirmesi gereken değişken olan "purchase_count"ı dönüştürmesi gerektiğini ifade ediyoruz.

df.head()
```

```python
df.describe().T
```

```python
# Aynı işlemi bir de "commment_count" değişkeni için yapmamız gerekmektedir;

df["comment_count_scaled"] = MinMaxScaler(feature_range=(1, 5)). \
    fit(df[["commment_count"]]). \
    transform(df[["commment_count"]])

df.head()
```

```python
# Şu anda bu skorların ortalamasını alabilir veya ağırlıklandırabiliriz;

df["weighted_sorting_score"] = (df["comment_count_scaled"] * 32 / 100 +
                                df["purchase_count_scaled"] * 26 / 100 +
                                df["rating"] * 42 / 100)

 
```

```python
df.sort_values("weighted_sorting_score", ascending=False)[["course_name", "instructor_name", "purchase_count", "rating", "commment_count", "weighted_sorting_score"]].head(20)
```

### Bayes Ortalama Derecelendirme Puanı ( Bayesian Average Rating Score )

```python
# Sorting Products with 5 Star Rated
# Sorting Products According to Distribution of 5 Star Rating
# isimleri ile de karşımıza çıkabilir.

df.head()
```

```python
# "Bayesian Average Rating" puanların dağılımı üzerinden (veriler 5, 4, 3 gibi puanlar) ağırlıklı bir şekilde olasılıksal ortalama hesaplar.

def bayesian_average_rating(n, confidence=0.95):
    if sum(n) == 0:
        return 0
    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n[k] + 1) / (N + K)
        second_part += (k + 1) * (k + 1) * (n[k] + 1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N + K + 1))
    return score

df["bar_score"] = df.apply(lambda x: bayesian_average_rating(x[["1_point",
                                                                "2_point",
                                                                "3_point",
                                                                "4_point",
                                                                "5_point"]]), axis=1)

df.head()
```
